```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2,2, figsize=(13, 10), sharey=True)

method = ["BLEU-1","BLEU-2","BLEU-3","BLEU-4"]

fg_vgg16 = [0.387293,0.186633,0.111938,0.04177]
fg_vgg19 = [0.295786,0.145646,0.088527,0.030013]
fg_resnet = [0.452356,0.241076,0.150023,0.059307]
fg_inception= [0.3503133,0.183449,0.116926,0.043951]
axs[0, 0].plot(method, fg_vgg16, label = 'VGG-16', color='red', linestyle='dashed')
axs[0, 0].plot(method, fg_vgg19, label = 'VGG-19', color='g', linestyle='solid')
axs[0, 0].plot(method, fg_resnet, label = 'Resnet', color='b', linestyle='dashed')
axs[0, 0].plot(method, fg_inception, label = 'Inception', color='m', linestyle='solid')
axs[0, 0].set_title('Flickr-GloVe : BLEU Score by Method')

fb_vgg16 = [0.433748,0.22528,0.139985,0.052553]
fb_vgg19 = [0.414702,0.225108,0.138914,0.052292]
fb_resnet = [0.481651,0.268223,0.168761,0.068185]
fb_inception= [0.341801,0.182212,0.110341,0.037636]
axs[0, 1].plot(method, fb_vgg16, label = 'VGG-16', color='red', linestyle='dashed')
axs[0, 1].plot(method, fb_vgg19, label = 'VGG-19', color='g', linestyle='solid')
axs[0, 1].plot(method, fb_resnet, label = 'Resnet', color='b', linestyle='dashed')
axs[0, 1].plot(method, fb_inception, label = 'Inception', color='m', linestyle='solid')
axs[0, 1].set_title('Flickr-BERT : BLEU Score by Method')

cg_vgg16 = [0.3875,0.161678,0.079887,0.024933]
cg_vgg19 = [0.366374,0.162469,0.087531,0.032032]
cg_resnet = [0.411474,0.194929,0.121182,0.048284]
cg_inception= [0.392499,0.204241,0.134606,0.060376]
axs[1, 0].plot(method, cg_vgg16, label = 'VGG-16', color='red', linestyle='dashed')
axs[1, 0].plot(method, cg_vgg19, label = 'VGG-19', color='g', linestyle='solid')
axs[1, 0].plot(method, cg_resnet, label = 'Resnet', color='b', linestyle='dashed')
axs[1, 0].plot(method, cg_inception, label = 'Inception', color='m', linestyle='solid')
axs[1, 0].set_title('COCO-GloVe : BLEU Score by Method')

cb_vgg16 = [0.391411,0.176105,0.087608,0.018199]
cb_vgg19 = [0.331898,0.141391,0.070898,0.018602]
cb_resnet = [0.396902,0.194672,0.11105,0.04008]
cb_inception= [0.361405,0.168333,0.09171,0.029009]
axs[1, 1].plot(method, cb_vgg16,   label = 'VGG-16', color='red', linestyle='dashed')
axs[1, 1].plot(method, cb_vgg19,  label = 'VGG-19', color='g', linestyle='solid')
axs[1, 1].plot(method, cb_resnet, label = 'Resnet', color='b', linestyle='dashed')
axs[1, 1].plot(method, cb_inception, label = 'Inception', color='m', linestyle='solid')
axs[1, 1].set_title('COCO-BERT : BLEU Score by Method')

# vertical lines for each BLEU score
for xc in method:
    axs[0, 0].axvline(x=xc, linewidth=.5)
    axs[0, 1].axvline(x=xc, linewidth=.5)
    axs[1, 0].axvline(x=xc, linewidth=.5)
    axs[1, 1].axvline(x=xc, linewidth=.5)

# set common yticks
plt.yticks(np.arange(0, 0.5, step=0.05))

# set x and y labels
for ax in axs.flat:
    ax.set(xlabel='Score Method', ylabel='Score')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()

# reduce whitespace between
plt.tight_layout() 

# add legends to each
axs[0, 0].legend(loc='upper right')
axs[0, 1].legend(loc='upper right')
axs[1, 0].legend(loc='upper right')
axs[1, 1].legend(loc='upper right')

plt.savefig('AllResultsGraph_HighResolution.png', dpi=300)
```
